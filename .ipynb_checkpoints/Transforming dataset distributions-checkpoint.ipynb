```python
# import libraries
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
# settings
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
# import the csv file and define columns of the dataset
address = 'demo/data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
# plot mpg variable
mpg = cars.mpg
mpg.plot()
```

```python
# generate a sumnmary of statistics from mpg variable
cars[['mpg']].describe()
```

```python
# Make sure mpg variable is a 1 column matrix
mpg_matrix = mpg.values.reshape(-1, 1) # 1 column matrix

# scale mpg variable within the range of 0 to 1
scaled = preprocessing.MinMaxScaler()
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
# expand the scale to 0 to 10

scaled = preprocessing.MinMaxScaler(feature_range=(0, 10))
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
standardized_mpg = scale(mpg, axis = 0, with_mean=False, with_std=False)
plt.plot(standardized_mpg)
```

```python
# standardize the variable 'mpg'
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg)
```
